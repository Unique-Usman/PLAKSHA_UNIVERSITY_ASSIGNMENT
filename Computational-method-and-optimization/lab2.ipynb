```python
#Question 1
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the symbol and the functions
x = sp.Symbol('x')
f_x = 1 / (1 - x**2)
g_x = 1 / (1 + x**2)

# Create a range of x values for plotting
gx_values = np.linspace(-2, 2, 400) 
fx_values = np.linspace(-2, 2, 100) 

# Convert the symbolic functions to Python functions for numerical evaluation
f_x_numeric = sp.lambdify(x, f_x, 'numpy')
g_x_numeric = sp.lambdify(x, g_x, 'numpy')

# Calculate and plot the Taylor series expansions
fn_terms = 5  # Number of terms to consider in the Taylor series
gn_terms = 4

# Initialize arrays to store the Taylor series approximations
f_taylor = np.zeros_like(fx_values)
g_taylor = np.zeros_like(gx_values)

for n in range(gn_terms):
    g_taylor_term = sp.series(g_x, x, 0, n).removeO()
    
    g_taylor += sp.lambdify(x, g_taylor_term, 'numpy')(gx_values)
for n in range(fn_terms):
    f_taylor_term = sp.series(f_x, x, 0, n).removeO()  
    
    f_taylor += sp.lambdify(x, f_taylor_term, 'numpy')(fx_values)

# Plot the functions and their Taylor series approximations
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.plot(fx_values, f_x_numeric(fx_values), label='f(x)')
plt.plot(fx_values, f_taylor, label='Taylor Expansion')
plt.title('Function f(x) and Taylor Series Expansion')
plt.legend()

plt.subplot(122)
plt.plot(gx_values, g_x_numeric(gx_values), label='g(x)')
plt.plot(gx_values, g_taylor, label='Taylor Expansion')
plt.title('Function g(x) and Taylor Series Expansion')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_square_lattice(num_circles, radius):
    fig, ax = plt.subplots()
    ax.set_aspect('equal')

    for i in range(num_circles):
        for j in range(num_circles):
            x = i * 2 * radius + radius
            y = j * 2 * radius + radius
            circle = plt.Circle((x, y), radius, fill=False, color='blue', linewidth=1)
            ax.add_patch(circle)

    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    plt.title(f'Square Lattice with {num_circles} Circles')
    plt.show()

def plot_hexagonal_lattice(num_circles, radius):
    fig, ax = plt.subplots()
    ax.set_aspect('equal')

    for i in range(num_circles):
        for j in range(num_circles):
            x = i * np.sqrt(3) * radius + (j % 2) * np.sqrt(3) * radius / 3
            y = j * 1.5 * radius
            circle = plt.Circle((x, y), radius, fill=False, color='red', linewidth=1)
            ax.add_patch(circle)

    ax.set_xlim(-radius, 10)
    ax.set_ylim(-radius, 10)
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    plt.title(f'Hexagonal Lattice with {num_circles} Circles')
    plt.show()

# Number of circles you want to use
num_circles = 20

# Radius of the circles
radius = 0.5 

# Plot square lattice
plot_square_lattice(num_circles, radius)

# Plot hexagonal lattice
plot_hexagonal_lattice(num_circles, radius)
```

Looking at the above shape, the hexagonal arrangement covers about 0.8 in area, while for the square arrangement, it covers about 1 in area of the total square for the same number of circles. The hexagonal arrangement is more optimal for space because it saves more space, which can be used to fit more circles. After varying the number of circles used, the ratio of the area covered for the square remained constant, while the hexagonal arrangement kept decreasing as it was able to accommodate more circles.
