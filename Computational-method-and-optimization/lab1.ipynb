```python
import numpy as np
import matplotlib.pyplot as plt
# Question 1
start = 0.00
stop = 0.02 + 0.001
step = 0.001
eps = np.arange(start, stop, step)
x = [0.9 / (1 - (101 * e )) for e in eps]
y = [(11 - 0.9) / ((1- 101 * e) * 101) for e in eps]
```

```python
# Create a new figure and axis
fig, ax = plt.subplots()

# Plot the two y-values against the x-value
ax.plot(eps, x, label='X Values')
ax.plot(eps, y, label='Y2 Values')

# Set labels for x-axis and y-axis
ax.set_xlabel('X Values')
ax.set_ylabel('Y Values')

# Add a legend
ax.legend()

#add a title
plt.title("The perturbation")

# Show the plot
plt.show()
```
